**ANALISE DE VENDAS,SERENATTO2024**

```python
import pandas as pd #importando a biblioteca pandas
```

```python
vendas = pd.read_csv("serenatto_2sem_2023.csv") #lendo o arquivo csv com os dados de vendas.
vendas.query('produto == "Tiramisù" ').head() #estamos filtrando o DataFrame para selecionar apenas as linhas em que o produto seja “Tiramisù”.
```

```python
vendas.query('valor > 10 and metodo_pagamento != "Dinheiro" ') #estamos filtrando o DataFrame para selecionar apenas as linhas em que o valor seja maior que 10 e o método de pagamento não seja “Dinheiro”.
```

```python
produtos = ['Café au lait', 'Espresso', 'Cappucino'] #Aqui estou  criando uma lista de produtos que queremos analisar.
vendas.query('produto in @produtos and metodo_pagamento == "PIX" ') #Aqui estou  filtrando o DataFrame para selecionar apenas as linhas em que o produto estiver na lista de produtos e o método de pagamento seja “PIX”.
```
