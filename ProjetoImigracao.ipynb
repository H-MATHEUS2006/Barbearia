# **Analise tendencia de imigração para BRASIL**

```python
df.set_index('País', inplace=True)
```

```python
anos = list(map(str, range(1980, 2014)))
```

```python
anos
```

```python
brasil = df.loc['Brasil', anos]
brasil
```

# **Aqui esta sendo criado um dataFrame para exibir o ano e nuemro de imigrante que veio para brasil**

```python
brasil_dict = {'ano': brasil.index.to_list(), 'imigrantes': brasil.values.tolist()}
dados_brasil = pd.DataFrame(brasil_dict)
dados_brasil
```

# **Criação do grafico para analise de imigrantes que veio para brasil**

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 4))
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
plt.title('Tendencia de imigrantes do BRASIL para CANADÁ')
plt.xlabel('Anos')
plt.ylabel('Numero de imigrantes')
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
plt.xlabel('Anos')
plt.ylabel('Numero de imigrantes')
plt.show()
#plt.yticks([500, 1000, 1500, 2000, 2500, 3000])
```

# ***Criando grafico com fig***

```python
fig, ax = plt.subplots(figsize=(10, 4))
ax.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.set_title('Tendencia de imigrantes do BRASIL para CANADÁ\n1980 á 2013')
ax.set_xlabel('Anos')
ax.set_ylabel('Numero de imigrantes')
plt.show()
```
